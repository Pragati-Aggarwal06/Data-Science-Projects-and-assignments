DATA SCIENCE PROJECT 1 (HOUSE SALES PREDICTION) MATPLOTLIB AND SEABORN.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("/content/House Sales Prediction.csv")
```

PART - A :- MATPLOTLIB QUESTIONS.

QUES:-1. How are house prices distributed in the dataset?

Plot: Histogram

```python
plt.hist(df['price'], bins=30)
plt.title("Distribution of House Prices")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()
```

Conclusion: Identifies whether prices are normally distributed or skewed.

QUES:-2. Does living area affect house price?

Plot: Scatter Plot

```python
plt.scatter(df['sqft_living'], df['price'])
plt.xlabel("Living Area")
plt.ylabel("Price")
plt.title("Living Area vs Price")
plt.show()
```

Conclusion: Larger homes generally have higher prices

QUES:-3. What is the average house price for different bedroom counts?

Plot: Bar Char

```python
avg_price = df.groupby('bedrooms')['price'].mean()

plt.bar(avg_price.index, avg_price.values)
plt.xlabel("Bedrooms")
plt.ylabel("Average Price")
plt.title("Average Price by Bedrooms")
plt.show()
```

QUES:-4. How does the number of bathrooms influence price?

```python
avg_price = df.groupby('bathrooms')['price'].mean()

plt.plot(avg_price.index, avg_price.values, marker='o')
plt.xlabel("Bathrooms")
plt.ylabel("Average Price")
plt.title("Bathrooms vs Price")
plt.show()
```

QUES:-5. Compare prices of waterfront and non-waterfront houses.

```python
water_price = df.groupby('waterfront')['price'].mean()

plt.bar(['No Waterfront','Waterfront'], water_price)
plt.title("Waterfront Impact on Price")
plt.show()
```

QUES:-6. Which cities have the highest average house prices?

```python
city_price = df.groupby('city')['price'].mean().sort_values(ascending=False).head(10)

city_price.plot(kind='bar')
plt.title("Top 10 Cities by Average Price")
plt.show()
```

QUES:-7. How has house construction changed over time?

```python
year_counts = df['yr_built'].value_counts().sort_index()

plt.plot(year_counts.index, year_counts.values)
plt.title("Houses Built Over Time")
plt.show()
```

QUES:-8. What is the distribution of house conditions?

```python
condition_counts = df['condition'].value_counts()

plt.pie(condition_counts,
        labels=condition_counts.index,
        autopct='%1.1f%%')
plt.show()
```

QUES:-9. Are houses with more floors more expensive?

```python
floor_price = df.groupby('floors')['price'].mean()

plt.bar(floor_price.index.astype(str), floor_price.values)
plt.show()
```

QUES:-10. Identify outliers in house prices.

```python
plt.boxplot(df['price'])
plt.title("House Price Outliers")
plt.show()
```

PART - B:- SEABORN QUESTIONS.

QUES:-1. What is the relationship between living area and price?

```python
sns.regplot(x='sqft_living', y='price', data=df)
plt.show()
```

Conclusion: Strong positive relationship.

QUES:-2. Which numerical features are most correlated with price?

```python
corr = df.corr(numeric_only=True)

sns.heatmap(corr, annot=True)
plt.show()
```

QUES:-3. Compare house prices across different conditions.

```python
sns.boxplot(x='condition', y='price', data=df)
plt.show()
```

QUES:-4. How does waterfront availability affect price distribution?

```python
sns.violinplot(x='waterfront', y='price', data=df)
plt.show()
```

QUES:-5. Which bedroom category is most common?

```python
sns.countplot(x='bedrooms', data=df)
plt.show()
```

QUES:-6. How are bathrooms distributed?

```python
sns.histplot(df['bathrooms'], kde=True)
plt.show()
```

QUES:-7. Compare price distributions among top cities.

```python
top_cities = df['city'].value_counts().head(10).index

sns.boxplot(
    x='city',
    y='price',
    data=df[df['city'].isin(top_cities)]
)

plt.xticks(rotation=45)
plt.show()
```

QUES:-8. Visualize relationships among important variables.

```python
sns.pairplot(df[['price',
                 'sqft_living',
                 'bedrooms',
                 'bathrooms',
                 'sqft_lot']])
plt.show()
```

QUES:-9. How does house view rating affect price?

```python
sns.barplot(x='view', y='price', data=df)
plt.show()
```

QUES:-10. Does house condition influence living area?

```python
sns.scatterplot(
    x='condition',
    y='sqft_living',
    hue='price',
    data=df
)
plt.show()
```

TASK 4:- DATA CLEANING.

4.1) Handle Missing Values.

```python
df.isnull().sum()
```

```python
df['director'] = df['director'].fillna('Unknown')
df['cast'] = df['cast'].fillna('Unknown')
df['country'] = df['country'].fillna('Unknown')
df['rating'] = df['rating'].fillna('Unknown')
df['date_added'] = df['date_added'].fillna('Unknown')
```

4.2) Remove Duplicates.

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

4.3) Correct Data Types.

```python
df['date_added'] = df['date_added'].replace('Unknown', np.nan)
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df.dtypes
```

TASK 5:- OUTLIER DETECTION.

5.1) Detect Outliers using a Boxplot.bold text

```python
plt.figure(figsize=(8,5))
sns.boxplot(x=df['release_year'])
plt.title("Boxplot of Release Year")
plt.xlabel("Release Year")
plt.show()
```

5.2) Handle Outliers using the IQR or Z-scoe method.

```python
Q1 = df['release_year'].quantile(0.25)
Q3 = df['release_year'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)
```

```python
df_no_outliers = df[(df['release_year'] >= lower_bound) & (df['release_year'] <= upper_bound)]

print("Original shape:", df.shape)
print("New shape after removing outliers:", df_no_outliers.shape)
```

TASK 6:- UNIVARIATE ANALYSIS.

Create suitable plots and analyze individual variables.

6.1) Count of Movies vs TV Shows.

```python
plt.figure(figsize=(6,4))
sns.countplot(x='type', data=df)
plt.title("Count of Movies vs TV Shows")
plt.xlabel("Type")
plt.ylabel("Count")
plt.show()
```

6.2) Distribution of Ratings.

```python
plt.figure(figsize=(12,6))
sns.countplot(y='rating', data=df, order=df['rating'].value_counts().index)
plt.title("Distribution of Content Ratings")
plt.xlabel("Count")
plt.ylabel("Rating")
plt.show()
```

6.3) Titles Released Per Year.

```python
plt.figure(figsize=(12,6))
sns.histplot(df['release_year'], bins=30, kde=True)
plt.title("Distribution of Release Year")
plt.xlabel("Release Year")
plt.ylabel("Frequency")
plt.show()
```

6.4) Top 10 Countries with Highest Content.

```python
top_countries = df[df['country'] != 'Unknown']['country'].value_counts().head(10)

plt.figure(figsize=(12,6))
sns.barplot(x=top_countries.values, y=top_countries.index)
plt.title("Top 10 Countries with Highest Netflix Content")
plt.xlabel("Number of Titles")
plt.ylabel("Country")
plt.show()
```

TASK 7:- BIVARIATE ANALYSIS.

Analyze relationships between two variables using appropriate visualizations.

7.1) Type vs Rating.

```python
plt.figure(figsize=(12,6))
sns.countplot(data=df, y='rating', hue='type', order=df['rating'].value_counts().index)
plt.title("Rating Distribution by Type")
plt.xlabel("Count")
plt.ylabel("Rating")
plt.show()
```

7.2) Type vs Release Year.

```python
plt.figure(figsize=(10,5))
sns.boxplot(x='type', y='release_year', data=df)
plt.title("Release Year by Content Type")
plt.xlabel("Type")
plt.ylabel("Release Year")
plt.show()
```

7.3) Country vs Type.

```python
top10_country_names = df[df['country'] != 'Unknown']['country'].value_counts().head(10).index
df_top_country = df[df['country'].isin(top10_country_names)]

plt.figure(figsize=(12,6))
sns.countplot(data=df_top_country, y='country', hue='type')
plt.title("Top Countries by Content Type")
plt.xlabel("Count")
plt.ylabel("Country")
plt.show()
```

7.4) Year-wise Addition of Content to Netflix.

```python
df['year_added'] = df['date_added'].dt.year

plt.figure(figsize=(12,6))
sns.countplot(x='year_added', data=df, order=sorted(df['year_added'].dropna().unique()))
plt.xticks(rotation=45)
plt.title("Number of Titles Added to Netflix by Year")
plt.xlabel("Year Added")
plt.ylabel("Count")
plt.show()
```

TASK 9:- FEATURE ENGINEERING.

Create at least one new meaningful feature.

Feature 1: Content Age.

```python
current_year = df['release_year'].max()
df['content_age'] = current_year - df['release_year']
df[['title', 'release_year', 'content_age']].head()
```

This tells how old the content is compared to the latest year in the dataset.

Feature 2: Year Added / Month Added

```python
df['year_added'] = df['date_added'].dt.year
df['month_added'] = df['date_added'].dt.month_name()
```

These are useful for trend analysis.
